```python
# Section 2: Define Reference Files and Preprocess Function

import torchaudio

accent_files = {
    "american": ["american_1.wav", "american_2.wav", "american_3.wav"],
    "british": ["british_1.wav", "british_2.wav", "british_3.wav"],
    "indian": ["indian_1.wav", "indian_2.wav", "indian_full.wav"],
    "nigerian": ["nigerian_1.wav", "nigerian_2.wav", "nigerian_full.wav"],
}

def preprocess_wav(src, dst):
    waveform, sr = torchaudio.load(src)
    if waveform.shape[0] > 1:
        waveform = waveform.mean(dim=0, keepdim=True)
    if sr != 16000:
        waveform = torchaudio.transforms.Resample(sr, 16000)(waveform)
    torchaudio.save(dst, waveform, 16000)
```

```python
# Section 3: Load the Model and Build Reference Embeddings (RAM Safe)

from transformers import Wav2Vec2Processor, Wav2Vec2Model
import torch
import numpy as np

processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base-960h")

def extract_short_embedding(file_path, duration_sec=10):
    waveform, sample_rate = torchaudio.load(file_path)
    max_samples = int(sample_rate * duration_sec)
    waveform = waveform[:, :max_samples]
    if waveform.shape[0] > 1:
        waveform = waveform.mean(dim=0, keepdim=True)
    if sample_rate != 16000:
        waveform = torchaudio.transforms.Resample(sample_rate, 16000)(waveform)
    inputs = processor(waveform.squeeze().numpy(), sampling_rate=16000, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
    return embedding

reference_embeddings = {}
for accent, files in accent_files.items():
    reference_embeddings[accent] = []
    for file in files:
        try:
            emb = extract_short_embedding(file, duration_sec=10)
            reference_embeddings[accent].append(emb)
        except Exception as e:
            print(f"Error processing {file}: {e}")

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
# Section 4: Predict Accent from a Video URL (End-to-End)

# 1. Set the video URL
video_url = "https://www.youtube.com/watch?v=8dGl_9Kk18c"

# 2. Download and extract audio
!yt-dlp -x --audio-format wav -o "input_audio.%(ext)s" "{video_url}"

# 3. Preprocess the new audio
preprocess_wav('input_audio.wav', 'input_audio.wav')

# 4. Extract embedding for the first 10 seconds
input_emb = extract_short_embedding('input_audio.wav', duration_sec=10)

# 5. Predict accent
scores = {}
for accent, embs in reference_embeddings.items():
    similarities = [cosine_similarity(input_emb, ref_emb) for ref_emb in embs]
    scores[accent] = np.mean(similarities) if similarities else float('-inf')

predicted_accent = max(scores, key=scores.get)
print("Predicted accent:", predicted_accent)
print("Scores by accent:", scores)
```
